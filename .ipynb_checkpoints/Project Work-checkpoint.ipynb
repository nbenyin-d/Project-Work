##

## Data Cleaning
---
###

### > Check missing values in data set

```python
dataset.isnull().sum()
```

```python
median_fa = data_stat.fixed_acidity.loc['50%']
median_va = data_stat.volatile_acidity.loc['50%']
median_c = data_stat.chlorides.loc['50%']
median_fsd = data_stat.free_sulfur_dioxide.loc['50%']
median_d = data_stat.density.loc['50%']
median_q = data_stat.quality.loc['50%']
```

```python
nData = dataset.copy()
```

```python
nData.fixed_acidity = dataset.fixed_acidity.fillna(median_fa)
nData.volatile_acidity = dataset.volatile_acidity.fillna(median_va)
nData.chlorides = dataset.chlorides.fillna(median_c)
nData.free_sulfur_dioxide = dataset.free_sulfur_dioxide.fillna(median_fsd)
nData.density = dataset.density.fillna(median_d)
nData.quality = dataset.quality.fillna(median_q)
```

```python
nData.isnull().sum()
```

### > Check duplicated values 

```python
nData.duplicated()
```

```python
nData =nData.drop_duplicates()
```

```python
nData.duplicated()
```

```python
nData_stat =nData.describe()
nData_stat
```

## Scatter graph; Fixed acidity against pH
---

```python
%matplotlib inline
```

```python
mat.figure(figsize= (8,6))
mat.title('Fixed acidity over pH', fontdict= {'fontweight': 'bold'})
mat.scatter(x='pH',y='fixed_acidity',data = nData, c='g', alpha= 0.6)
mat.xlabel('pH')
mat.ylabel('Fixed acidity')

mat.legend()
```

### From the scatter graph above, the fixed acidity of each red wine was plotted against it's pH. By observing the graph, it can be deduced that those with a lower pH had a higher fixed acidity compared to those with higher pH.
### This gives the graph above something like a negative slope.Therefore, it can be said that the higher pH of red wine the lower its fixed acidity.

## Histogram; with respect to Density
---

```python
bins = np.arange(0.990,1.003,0.001)

mat.figure(figsize= (7,5))
mat.hist(nData.density, bins= bins, rwidth= 0.6)
mat.xticks(bins)

mat.title('Distribution of densities of Red Wine', fontdict= {'fontweight': 'bold'})
mat.xlabel('Density')
mat.ylabel('Red wine')

mat.show()
```

### From the histogram, most red wines possess densities within the range of  0.996 - 0.997 .
### At the same time few red wines possess densities within the range of 0.990 - 0.991 .

## Pie Chart
---

#### > The average amount of citric acid, residual sugar, chlorides, free sulfur dioxide, sulphates and alcohol in red wine.

```python
nData_stat[['citric_acid','residual_sugar','chlorides','free_sulfur_dioxide', 'sulphates','alcohol']].loc['mean']
```

```python
citric_acid =nData_stat['citric_acid'].loc['mean']
residual_sugar =nData_stat['residual_sugar'].loc['mean']
chlorides =nData_stat['chlorides'].loc['mean']
free_sulfur_dioxide =nData_stat['free_sulfur_dioxide'].loc['mean']
sulphates =nData_stat['sulphates'].loc['mean']
alcohol =nData_stat['alcohol'].loc['mean']

labels = ['Citric acid','Residual sugar','Chlorides','Free sulfur dioxide','Sulphates','Alcohol']
explode_values = [0.2,0,0.3,0,0.1,0]

mat.pie([citric_acid, residual_sugar, chlorides, free_sulfur_dioxide, sulphates, alcohol], labels = labels, autopct= '%.2f %%',explode=explode_values)
mat.title('Components of Red Wine', fontdict= {'fontweight': 'bold'})

mat.show()
```

### Even though the respective components' quantities of different red wines were provided, the mean  of each component quantity was used.
### This is to make it seem like there is only one red wine as we compare the componets to help us determine the percentage in  red wine.
>NB: **_This is only an average estimation. Also dimensional units are not taken into consideration._**

| Component | Percentage in Red Wine |
| --- | --- |

| Residual sugar | 8.51 % |
| Chlorides | 0.29 % |

## Classifying red wine baded on it's quality.
---

```python
def rating(rating):
    if rating >= 7.0:
        return 'Excellent'
    elif rating >= 6.0:
        return 'good'
    elif rating <= 3.0:
        return 'poor'
    else:
        return 'average'
```

```python
new_set = nData.copy()

new_set['wine_rating']= new_set.quality.apply(rating)
```
